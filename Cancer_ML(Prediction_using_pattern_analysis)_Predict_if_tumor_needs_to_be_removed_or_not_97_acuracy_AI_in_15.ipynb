```python
import pandas as pd

dataset = pd.read_csv('cancer.csv')
```

```python

# x (the data which the ml learning model will analyse using a certain standard pattern to predict y). The accuracy of the model will be based on the accuracy of the pattern.

x = dataset.drop(columns=["diagnosis(1=m, 0=b)"])
```

```python
# y is the value we want the ml model to predict using x (the data which the ml learning model will analyse using a certain standard pattern to predict y). 
# The accuracy of the model will be based on the accuracy of the pattern.

y = dataset["diagnosis(1=m, 0=b)"]
```

Split the data into a training set and a testing set.

```python
# The train and test data is split into 80% and 20% respectively. 
# The train data (80%) is used to train the model and the test data (20%) is used to test the accuracy of the model.

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

Build and train the model.

```python
import tensorflow as tf

model = tf.keras.models.Sequential()
```

```python
# Image Copy below image from:
# https://en.wikipedia.org/wiki/Sigmoid_function

from IPython.display import Image
Image(filename='320px-Logistic-curve.svg.png')
```

```python

# model.add(tf.keras.layers.Dense(256, input_shape=x_train.shape, activation='sigmoid'))
# model.add(tf.keras.layers.Dense(256, input_shape=x_train.shape[1:], activation='sigmoid'))


# What does the sigmoid function return? 
# Source:https://en.wikipedia.org/wiki/Sigmoid_function

# text below from:
# https://en.wikipedia.org/wiki/Sigmoid_function

# The activator function is used to determine the output of a neuron given a set of inputs.
# The activator function is Sigmoid.
# Sigmoid functions have domain of all real numbers, with return (response) value commonly monotonically increasing but could be decreasing. 
# Sigmoid functions most often show a return value (y axis) in the range 0 to 1. Another commonly used range is from âˆ’1 to 1.

model.add(tf.keras.layers.Dense(256, input_shape=(x_train.shape[1],), activation='sigmoid'))
model.add(tf.keras.layers.Dense(256, activation='sigmoid'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=1000)
```

Evaluate the model.

```python
model.evaluate(x_test, y_test)
```
